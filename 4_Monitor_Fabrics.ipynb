## Create and monitor a Dataflow

1. from the lakehouse -> get new data -> DataFlow (Gen2)

2. left pane "Monitor" to check creation of table in the lakehouse

3. verify table were created

## Create and monitor a Spark notebook

1. left pane "..." notebook

2. Explorer pane-> Add data items -> Existing data sources -> select lakehouse

3. Explorer pane -> Table -> Load Data -> Spark

4. left pane "Monitor" to check activity



## Monitor history for an item


1. in the … menu for the Get Product Data dataflow, select Historical run

2. In the … menu for any of the historical runs select View detail

## create eventstream

## create eventhouse
1. your eventhouse contains a KQL database, select Get data.
2. For the data source, select Eventstream > Existing eventstream.
3. Select or create a destination table 

## query database with KQL

```python
 stock
 | where ["time"] > ago(5m)
 | summarize avgPrice = avg(todecimal(bidPrice)) by symbol
 | project symbol, avgPrice
```

## Create a real-time dashboard
select the KQL query you used to retrieve the average stock prices for the last five minutes.
On the toolbar, select Save to dashboard

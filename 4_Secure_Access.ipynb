## set up

1. create a workspace
2. create a warehouse
3. create a lakehouse (-> add data )

# workspace access

1. in worksapce -> manage access
assign the Workspace Viewer 
--> can now see all of the items in the workspace because they were assigned the Workspace Viewer role.
--> as a member of the Workspace Viewer role, you have CONNECT and ReadData permission on tables in the warehouse.

## Item permissiopn

1. "..." on the warehouse -> manage permission
2. check only "Read all data using SQL (ReadData) "

3. Use cant see the worksapce where the warehouse is 
4. find it with OneLake catalog on the left pane

## OneLake data access **roles** in a Lakehouse

1. from lakehouse -> manage onelake security

2. Create a new role 

3. select Selected data and Browse Lakehouse to find table to grant read access
